```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
arr = np.random.randint(20000,40000,100)
print(arr)
```

```python
#lineplot : only works on numerical data
arr.sort()      #phle array sort kr diye fr usse graph mein dala
plt.plot(arr)   #output mei '[<matplotlib.lines.Line2D at 0x136221b8550>]' isse hatane ke liye ye show use krte hai
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,10,9,30,45]
plt.plot(x,y)
plt.show()
```

```python
#jo pints hai unpe bullet points banne ke liye use marker = 'o'
x = [1,2,3,4,5]
y = [1,10,9,30,45]
plt.plot(x,y,marker = 'o')
plt.show()
''' 
| Marker | Description     |               |
| ------ | --------------- | ------------- |
| `'o'`  | Circle          |               |
| `'.'`  | Point           |               |
| `','`  | Pixel           |               |
| `'x'`  | X               |               |
| `'+'`  | Plus            |               |
| `'^'`  | Triangle Up     |               |
| `'v'`  | Triangle Down   |               |
| `'>'`  | Triangle Right  |               |
| `'<'`  | Triangle Left   |               |
| `'s'`  | Square          |               |
| `'p'`  | Pentagon        |               |
| `'*'`  | Star            |               |
| `'h'`  | Hexagon1        |               |
| `'H'`  | Hexagon2        |               |
| `'D'`  | Diamond         |               |
| `'d'`  | Thin Diamond    |               |
| \`'    | '\`             | Vertical Line |
| `'_'`  | Horizontal Line |               |

'''
```

```python
#ms=marker size,by default 6 hota hai, mfc=marker face color : circle ke andr ka color
#mec = marker edge color:circle ka outer color, color=black : line ka color black hoga
#lw = line width by defaulr 1 hoti hai, ls = line style
#alpha = transparency level:always lies between 0 and 1
#label = "lineplot" with plt.legend() :Graph ke corner me ek chhoti sa box aayega jisme lineplot likha hoga.



x = [1,2,3,4,5]
y = [1,10,9,30,45]
plt.plot(x,y,marker = 'o',ms = 10,             
         mfc = "red",mec = "green",
         color = "black", lw = 3,ls = "-",alpha = 1,label = "lineplot")
plt.title("plotting")               #graph ka title lgta hai
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

```python
#scatterplot: it also works on numerical data
plt.scatter(x,y,marker = 'd')
plt.show()

```

```python
#Bar chart
x = ["mp","up","ap","hr"]
y = [20,40,60,80]
colors = ["red","blue","green","yellow"]
hatch_patterns = ["//","\\\\","xx","oo"]      #bars ke andr ke patterns change krne ke liye
#plt.bar(x,y,color=colors,hatch = hatch_patterns,label = x,width = 0.6, edgecolor = "black", alpha = 1,linewidth = 3)
plt.barh(x,y)                                 #horizontal graph banane ke liye
plt.legend()
plt.show()
```

```python
#histogram : distributen of data in intervals(bins)
#bins=20 se intervals ki value change kr skte hai
arr = np.random.randint(20000,40000,100)
plt.hist(arr,bins=40)
plt.show()
```

```python
#pie chart
x = ["mp","up","ap","hr"]
y = [10,20,30,40]
explodes = (0.5,0,0,0)    
plt.pie(y,labels = x,autopct="%1.2f%%",shadow="true",explode = explodes)      #plotting y ke respect mein krenge
plt.savefig("piechart.jpg")
plt.show()
```

```python
#subplots
import warnings
warnings.filterwarnings("ignore")   #to ignore warning
x = np.linspace(0,10,30)
plt.figure(figsize=(10,6))
plt.subplot(2,2,1)        #2 rows,2 columns,1st graph
plt.plot(x,np.sin(x))
plt.title("sin")
plt.subplot(2,2,2)
plt.plot(x,np.cos(x))
plt.title("cos")
plt.subplot(2,2,3)
plt.plot(x,np.tan(x))
plt.title("tan")
plt.subplot(2,2,4)
plt.plot(x,np.log(x))
plt.title("log")
plt.show()





```
