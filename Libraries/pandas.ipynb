```python
#CLEANING DATA
#duplicated : check duplicates----normally true false mein deta hai answer
df.duplicated().sum()    #sum laga dene se woh batayega ki total kitne duplicates hai
df.drop_duplicates()     #duplicates delete krne ke liye
df.duplicated().sum()    #uske baad bhi show kr raha hai duplicates
df.drop_duplicates(inplace=True) #ye argument dene se jo chnages humne kiye hai duplicates mein woh save ho jayege
df.duplicated().sum() 

```

```python
#DATA CLEANING
#removing null values
df.isnull().sum()
df.dropna(inplace=True)  #isme inplace dena hi hoga
df.isnull().sum()
```
